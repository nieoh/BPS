### get recommendations

```python
def match(prediction_df, user_name, threshold=3):
    user_row = prediction_df.loc[user_name]
    weight = user_row[user_name]
    weighted_sorted_user_row = 1.00/weight * user_row.sort_values(ascending=False)
    return weighted_sorted_user_row[1:threshold+1]
```

```python
match(pred_df, 'Leo Dorest')
```

### Get headers

```python
headers_df = pd.read_csv('survey.csv', header=0, nrows=1)
```

```python
headers_df.dropna(axis=1,inplace=True)
```

```python
headers_df.drop([u'Unnamed: 1', u'Unnamed: 2', u'Unnamed: 3', u'Unnamed: 4',
       u'Unnamed: 5', u'Unnamed: 6', u'Unnamed: 7', u'Unnamed: 8',
       u'Unnamed: 9', u'Unnamed: 10', u'Unnamed: 12'], axis=1, inplace=True)
```

```python
col_name_dict = {}

for col in headers_df.columns:
    val = headers_df[col].iloc[0]
    if val in col_name_dict:
        print 'conflict!, {}, {}'.format(val, col)
    else:
        
        col_name_dict[str(val)] = col
```

```python
new_df_cols = []
for col in list(new_df.columns):
    if col in col_name_dict:
        new_df_cols.append(col_name_dict[col])
    else:
        new_df_cols.append(col)
```
